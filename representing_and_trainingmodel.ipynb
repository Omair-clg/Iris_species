```python
import matplotlib.pyplot as plt
from sklearn import datasets
iris= datasets.load_iris()
print(iris.keys())

features = iris.data.T
sepal_lengths = features[0]
sepal_width = features[1]
petal_lengths = features[2]
petal_width = features[3]

sepal_lengths_label = iris.feature_names[0]
sepal_width_label = iris.feature_names[1]
petal_lengths_label = iris.feature_names[2]
petal_width_label = iris.feature_names[3]

plt.scatter(sepal_lengths, sepal_width , c= iris.target)
plt.xlabel(sepal_lengths_label)
plt.ylabel(sepal_width_label)
plt.show()

plt.scatter(petal_lengths, petal_width , c= iris.target)
plt.xlabel(petal_lengths_label)
plt.ylabel(petal_width_label)
plt.show()
```

```python
import pandas as pd
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import joblib

iris = datasets.load_iris()

X_train, X_test, y_train, y_test = train_test_split(iris['data'], iris['target'], test_size=0.2, random_state=70)


model = DecisionTreeClassifier()
model.fit(X_train,y_train)

y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test,y_pred)
print(f"Model Accuracy: {accuracy:.2f}")


joblib.dump(model, 'Iris_DecisionTreeModel.pkl')

```

```python
import pandas as pd
from sklearn import datasets
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import numpy as np
import joblib

iris = datasets.load_iris()

X_train, X_test, y_train, y_test = train_test_split(iris['data'], iris['target'], test_size=0.2, random_state=50)


knn = KNeighborsClassifier(n_neighbors=1)

knn.fit(X_train, y_train)

y_pred = knn.predict(X_test)
accuracy = accuracy_score(y_test,y_pred)
print(f"Model Accuracy: {accuracy:.2f}")


joblib.dump(knn, 'Iris_KNeighborsClassifierModel.pkl')

```
